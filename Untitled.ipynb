```python
import pyspark.sql.functions as F
from pyspark.sql.types import *

# Define the schema matching the produce.py output
taxi_schema = StructType([
    StructField("VendorID", IntegerType(), True),
    StructField("tpep_pickup_datetime", StringType(), True), 
    StructField("tpep_dropoff_datetime", StringType(), True),
    StructField("passenger_count", DoubleType(), True),
    StructField("trip_distance", DoubleType(), True),
    StructField("PULocationID", IntegerType(), True),
    StructField("DOLocationID", IntegerType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("total_amount", DoubleType(), True),
    StructField("payment_type", IntegerType(), True),
    StructField("congestion_surcharge", DoubleType(), True)
])

# Load and prepare the static zone lookup table for the Silver layer join
zones = spark.read.parquet("data/taxi_zone_lookup.parquet")

pu_zones = zones.withColumnRenamed("LocationID", "PULocationID") \
                .withColumnRenamed("Zone", "pickup_zone") \
                .withColumnRenamed("Borough", "pickup_borough") \
                .select("PULocationID", "pickup_zone", "pickup_borough")

do_zones = zones.withColumnRenamed("LocationID", "DOLocationID") \
                .withColumnRenamed("Zone", "dropoff_zone") \
                .withColumnRenamed("Borough", "dropoff_borough") \
                .select("DOLocationID", "dropoff_zone", "dropoff_borough")
```

```python
# 1. Read from Kafka
raw_stream = (
    spark.readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "kafka:9092")
    .option("subscribe", "taxi-trips")
    .option("startingOffsets", "earliest")
    .load()
)

# 2. Extract raw strings and metadata
bronze_df = raw_stream.selectExpr(
    "CAST(key AS STRING) AS vendor_id",
    "CAST(value AS STRING) AS raw_json",
    "timestamp AS kafka_timestamp",
    "partition",
    "offset"
)

# 3. Write to Bronze Iceberg Table
bronze_query = (
    bronze_df.writeStream
    .format("iceberg")
    .outputMode("append")
    .queryName("bronze_stream")
    .trigger(processingTime="10 seconds")
    .option("checkpointLocation", "checkpoints/bronze")
    .toTable("lakehouse.taxi.bronze")
)
```

```python
# 1. Read from Bronze - Force it to pick up all existing data
silver_input_stream = (
    spark.readStream
    .format("iceberg")
    .option("streaming-skip-delete-snapshots", "true")
    .load("lakehouse.taxi.bronze")
)

# 2. Parse and Cast
parsed_df = silver_input_stream.select(
    F.from_json(F.col("raw_json"), taxi_schema).alias("data"),
    F.col("kafka_timestamp")
).select("data.*", "kafka_timestamp")

silver_df = parsed_df.withColumn(
    "pickup_datetime", F.to_timestamp("tpep_pickup_datetime")
).withColumn(
    "dropoff_datetime", F.to_timestamp("tpep_dropoff_datetime")
)

# 3. Clean and Join
# Using a very large watermark (500 days) so our 2025 data isn't dropped in 2026
enriched_df = silver_df.filter(
    (F.col("fare_amount") > 0) & 
    (F.col("trip_distance") >= 0) & 
    (F.col("PULocationID").isNotNull())
).join(pu_zones, on="PULocationID", how="left") \
 .join(do_zones, on="DOLocationID", how="left") \
 .withWatermark("pickup_datetime", "500 days") \
 .dropDuplicates(["VendorID", "pickup_datetime", "PULocationID"])

# 4. Write to Silver Table
silver_query = (
    enriched_df.writeStream
    .format("iceberg")
    .outputMode("append")
    .queryName("silver_stream")
    .trigger(processingTime="10 seconds")
    .option("checkpointLocation", "checkpoints/silver")
    .toTable("lakehouse.taxi.silver")
)
```

```python
# 1. Read from Silver
silver_stream = spark.readStream.format("iceberg").load("lakehouse.taxi.silver")

# 2. Aggregate
# We use a massive watermark here as well to ensure 2025 data is processed
gold_df = (
    silver_stream
    .withWatermark("pickup_datetime", "500 days") 
    .groupBy(
        F.window("pickup_datetime", "1 hour"),
        "pickup_zone"
    )
    .agg(
        F.count("*").alias("trip_count"),
        F.avg("fare_amount").alias("avg_fare"),
        F.sum("total_amount").alias("total_revenue")
    )
    .select(
        F.col("window.start").alias("window_start"),
        F.col("window.end").alias("window_end"),
        "pickup_zone",
        "trip_count",
        "avg_fare",
        "total_revenue"
    )
)

# 3. Write to Gold Table
gold_query = (
    gold_df.writeStream
    .format("iceberg")
    .outputMode("append") 
    .queryName("gold_stream")
    .trigger(processingTime="20 seconds")
    .option("checkpointLocation", "checkpoints/gold")
    .toTable("lakehouse.taxi.gold")
)
```

```python
# ── 2. STATIC DATA & SCHEMAS ─────────────────────────────────────────
BOOTSTRAP = "kafka:9092"
TOPIC     = "taxi-trips"

taxi_schema = StructType([
    StructField("VendorID", IntegerType(), True),
    StructField("tpep_pickup_datetime", StringType(), True), 
    StructField("tpep_dropoff_datetime", StringType(), True),
    StructField("passenger_count", DoubleType(), True),
    StructField("trip_distance", DoubleType(), True),
    StructField("PULocationID", IntegerType(), True),
    StructField("DOLocationID", IntegerType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("total_amount", DoubleType(), True),
    StructField("payment_type", IntegerType(), True),
    StructField("congestion_surcharge", DoubleType(), True)
])

zones = spark.read.parquet("data/taxi_zone_lookup.parquet")
pu_zones = zones.withColumnRenamed("LocationID", "PULocationID") \
                .withColumnRenamed("Zone", "pickup_zone") \
                .withColumnRenamed("Borough", "pickup_borough") \
                .select("PULocationID", "pickup_zone", "pickup_borough")

do_zones = zones.withColumnRenamed("LocationID", "DOLocationID") \
                .withColumnRenamed("Zone", "dropoff_zone") \
                .withColumnRenamed("Borough", "dropoff_borough") \
                .select("DOLocationID", "dropoff_zone", "dropoff_borough")

print("done")
```

```python
# ── 3. BRONZE LAYER ──────────────────────────────────────────────────
raw_stream = (
    spark.readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", BOOTSTRAP)
    .option("subscribe", TOPIC)
    .option("startingOffsets", "earliest")
    .load()
)

bronze_df = raw_stream.selectExpr(
    "CAST(key AS STRING) AS vendor_id",
    "CAST(value AS STRING) AS raw_json",
    "timestamp AS kafka_timestamp",
    "partition",
    "offset"
)

bronze_query = (
    bronze_df.writeStream
    .format("iceberg")
    .outputMode("append")
    .queryName("bronze_stream")
    .trigger(processingTime="10 seconds")
    .option("checkpointLocation", "/home/jovyan/checkpoints/bronze")
    .toTable("lakehouse.taxi.bronze")
)

print("done")
```

```python
# ── 4. SILVER LAYER ──────────────────────────────────────────────────
silver_input_stream = (
    spark.readStream
    .format("iceberg")
    .option("streaming-skip-delete-snapshots", "true")
    .load("lakehouse.taxi.bronze")
)

parsed_df = silver_input_stream.select(
    F.from_json(F.col("raw_json"), taxi_schema).alias("data"),
    F.col("kafka_timestamp")
).select("data.*", "kafka_timestamp")

silver_df = parsed_df.withColumn(
    "pickup_datetime", F.to_timestamp("tpep_pickup_datetime")
).withColumn(
    "dropoff_datetime", F.to_timestamp("tpep_dropoff_datetime")
)

# Custom Scenario Addition: is_weekend (1 = Sunday, 7 = Saturday)
enriched_df = silver_df.filter(
    (F.col("fare_amount") > 0) & 
    (F.col("trip_distance") >= 0) & 
    (F.col("PULocationID").isNotNull())
).join(pu_zones, on="PULocationID", how="left") \
 .join(do_zones, on="DOLocationID", how="left") \
 .withColumn("is_weekend", F.dayofweek("pickup_datetime").isin([1, 7])) \
 .withWatermark("pickup_datetime", "500 days") \
 .dropDuplicates(["VendorID", "pickup_datetime", "PULocationID"])

silver_query = (
    enriched_df.writeStream
    .format("iceberg")
    .outputMode("append")
    .queryName("silver_stream")
    .trigger(processingTime="10 seconds")
    .option("checkpointLocation", "/home/jovyan/checkpoints/silver")
    .toTable("lakehouse.taxi.silver")
)

print("done")
```

```python
# ── 5. GOLD LAYER ────────────────────────────────────────────────────
silver_stream = spark.readStream.format("iceberg").load("lakehouse.taxi.silver")

# Custom Scenario Addition: Added is_weekend to groupBy, avg_total_amount to agg
gold_df = (
    silver_stream
    .withWatermark("pickup_datetime", "500 days") 
    .groupBy(
        F.window("pickup_datetime", "1 hour"),
        "pickup_zone",
        "is_weekend"
    )
    .agg(
        F.count("*").alias("trip_count"),
        F.avg("fare_amount").alias("avg_fare"),
        F.sum("total_amount").alias("total_revenue"),
        F.avg("total_amount").alias("avg_total_amount")
    )
    .select(
        F.col("window.start").alias("window_start"),
        F.col("window.end").alias("window_end"),
        "pickup_zone",
        "is_weekend",
        "trip_count",
        "avg_fare",
        "total_revenue",
        "avg_total_amount"
    )
)

gold_query = (
    gold_df.writeStream
    .format("iceberg")
    .outputMode("append") 
    .queryName("gold_stream")
    .trigger(processingTime="20 seconds")
    .option("checkpointLocation", "/home/jovyan/checkpoints/gold")
    .toTable("lakehouse.taxi.gold")
)

print("done")
```

```python
import os
import shutil
import time
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import *

# ─── 1. HARD RESET (Clean Slate) ──────────────────────────────────────
if 'spark' in locals():
    for q in spark.streams.active:
        q.stop()
    spark.stop()

# Wipe checkpoints completely
for layer in ["bronze", "silver", "gold"]:
    path = f"/home/jovyan/checkpoints/{layer}"
    if os.path.exists(path):
        shutil.rmtree(path)

# ─── 2. SPARK SETUP ───────────────────────────────────────────────────
spark = (
    SparkSession.builder
    .appName("project2_final_victory")
    .master("local[*]")
    .config("spark.sql.shuffle.partitions", "4")
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions")
    .config("spark.sql.catalog.lakehouse", "org.apache.iceberg.spark.SparkCatalog")
    .config("spark.sql.catalog.lakehouse.type", "rest")
    .config("spark.sql.catalog.lakehouse.uri", "http://iceberg-rest:8181")
    .config("spark.sql.catalog.lakehouse.warehouse", "s3://warehouse/")
    .config("spark.sql.catalog.lakehouse.io-impl", "org.apache.iceberg.aws.s3.S3FileIO")
    .config("spark.sql.catalog.lakehouse.s3.endpoint", "http://minio:9000")
    .config("spark.sql.catalog.lakehouse.s3.path-style-access", "true")
    .config("spark.sql.catalog.lakehouse.s3.access-key-id", os.environ["AWS_ACCESS_KEY_ID"])
    .config("spark.sql.catalog.lakehouse.s3.secret-access-key", os.environ["AWS_SECRET_ACCESS_KEY"])
    .config("spark.sql.catalog.lakehouse.s3.region", "us-east-1")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")

# Drop tables to clear any corrupted schema locks
spark.sql("DROP TABLE IF EXISTS lakehouse.taxi.bronze")
spark.sql("DROP TABLE IF EXISTS lakehouse.taxi.silver")
spark.sql("DROP TABLE IF EXISTS lakehouse.taxi.gold")
spark.sql("CREATE DATABASE IF NOT EXISTS lakehouse.taxi")

# ─── 3. STATIC SCHEMAS & LOOKUPS ──────────────────────────────────────
taxi_schema = StructType([
    StructField("VendorID", IntegerType(), True),
    StructField("tpep_pickup_datetime", StringType(), True), 
    StructField("tpep_dropoff_datetime", StringType(), True),
    StructField("passenger_count", DoubleType(), True),
    StructField("trip_distance", DoubleType(), True),
    StructField("PULocationID", IntegerType(), True),
    StructField("DOLocationID", IntegerType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("total_amount", DoubleType(), True),
    StructField("payment_type", IntegerType(), True),
    StructField("congestion_surcharge", DoubleType(), True)
])

zones = spark.read.parquet("data/taxi_zone_lookup.parquet")
pu_zones = zones.select(F.col("LocationID").alias("PULocationID"), F.col("Zone").alias("pickup_zone"))
do_zones = zones.select(F.col("LocationID").alias("DOLocationID"), F.col("Zone").alias("dropoff_zone"))

# ─── 4. THE MASTER STREAM (Read Kafka Once) ───────────────────────────
raw_stream = (
    spark.readStream.format("kafka")
    .option("kafka.bootstrap.servers", "kafka:9092")
    .option("subscribe", "taxi-trips")
    .option("startingOffsets", "earliest")
    .load()
)

# --- BRONZE LOGIC ---
bronze_df = raw_stream.selectExpr(
    "CAST(key AS STRING) AS vendor_id",
    "CAST(value AS STRING) AS raw_json",
    "timestamp AS kafka_timestamp"
)

# --- SILVER LOGIC (Directly from Bronze DF) ---
parsed_df = bronze_df.select(F.from_json(F.col("raw_json"), taxi_schema).alias("data"), F.col("kafka_timestamp")).select("data.*", "kafka_timestamp")

silver_df = parsed_df.withColumn("pickup_datetime", F.to_timestamp("tpep_pickup_datetime")) \
                     .withColumn("dropoff_datetime", F.to_timestamp("tpep_dropoff_datetime")) \
                     .filter((F.col("fare_amount") > 0) & (F.col("PULocationID").isNotNull())) \
                     .join(pu_zones, on="PULocationID", how="left") \
                     .join(do_zones, on="DOLocationID", how="left") \
                     .withColumn("is_weekend", F.dayofweek("pickup_datetime").isin([1, 7])) \
                     .withWatermark("pickup_datetime", "10 minutes") \
                     .dropDuplicates(["VendorID", "pickup_datetime", "PULocationID"])

# --- GOLD LOGIC (Directly from Silver DF) ---
gold_df = silver_df.groupBy(F.window("pickup_datetime", "1 hour"), "pickup_zone", "is_weekend") \
                   .agg(F.count("*").alias("trip_count"), F.avg("total_amount").alias("avg_total_amount")) \
                   .select(F.col("window.start").alias("window_start"), "pickup_zone", "is_weekend", "trip_count", "avg_total_amount")


# ─── 5. START THE WRITE STREAMS ───────────────────────────────────────
print("Starting Bronze...")
bronze_query = bronze_df.writeStream.format("iceberg").outputMode("append").queryName("bronze_stream").trigger(processingTime="10 seconds").option("checkpointLocation", "/home/jovyan/checkpoints/bronze").toTable("lakehouse.taxi.bronze")

print("Starting Silver...")
silver_query = silver_df.writeStream.format("iceberg").outputMode("append").queryName("silver_stream").trigger(processingTime="10 seconds").option("checkpointLocation", "/home/jovyan/checkpoints/silver").toTable("lakehouse.taxi.silver")

print("Starting Gold...")
# Note: Gold partition strategy explicitly defined via Spark SQL before writing is best, 
# but Spark Structured Streaming auto-creates the table. We will let it auto-create.
gold_query = gold_df.writeStream.format("iceberg").outputMode("append").queryName("gold_stream").trigger(processingTime="20 seconds").option("checkpointLocation", "/home/jovyan/checkpoints/gold").toTable("lakehouse.taxi.gold")

time.sleep(10)
print("\n--- PIPELINE STATUS ---")
for q in spark.streams.active:
    print(f"✅ {q.name} is ACTIVE!")
```
